What can you infer about the relationships between (1) the women, and (2) the social events?

```python
import networkx as nx
import matplotlib.pyplot as plt
import networkx.algorithms.bipartite as bipartite
```

### Loading the data
Since the Davis Women graph is one that is preloaded with networkx, we will just load it in that way.

```python
G = nx.davis_southern_women_graph()
```

splitting the data.

```python
women = G.graph["top"]
clubs = G.graph["bottom"]
```

Now we will look at the matrices.

```python
print("Biadjacency matrix")
print(bipartite.biadjacency_matrix(G, women, clubs))
```

Now we want to visualize the graph.

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
nx.degree(G)
```

So we see that event 8 was the most popular event, and that 3 seperate women are tied for going to the most events.

```python
eigen = nx.eigenvector_centrality(G)
{k:v for k,v in sorted(eigen.items(), key=lambda item: item[1])}
```

There are a few things to note here. First, something important happened at event 8. Second, Brenda Rogers is a grey cardinal. She has a low amount of degrees, yet she has the 8th highest eigenvector score. Third, by looking at our original graph, we can maybe start assuming that Theresa Anderson and Evelyn Jefferson might be the leaders of the same clique, while Brenda and Slyvia are the leaders of the other clique.

```python
nx.betweenness_centrality(G)
```

Here we see that events 8 and 9 acted as bottlenecks between the nodes. Meaing these 2 events brought a lot of women out that might be in distinct groups.

Now we want to project each node type respectively.

We will begin by projecting the women nodes.

```python
projected_women = bipartite.projected_graph(G, women)
nx.draw(projected_women, with_labels=True)
plt.show()
```

Now we can run centrality metrics on this graph to find things about the southern women.

```python
nx.degree(projected_women)
```

Here we see that Evelyn, Theresa, Ruth, Verne, Sylvia, Nora, and Helen have been in the same room as everyone else at least once. This means that there is a high chance that they know all of the other women and that they themselves might serve as bridges between other women.

```python
eigen = nx.eigenvector_centrality(projected_women)
{k:v for k,v in sorted(eigen.items(), key=lambda item: item[1])}
```

We shouldn't be surprised to see that the same 7 women have the highest eigenvector centrality scores. It is slightly interesting that they are not equal scores, but the difference is not by much so we shouldn't be too worried. We should also note that the projection makes it seem like Brenda is not a grey cardinal.

```python
nx.closeness_centrality(projected_women)
```

Once again, we see the same 7 women have the highest scores in another metric. 

From analyzing the projection of just the women, we see that there appears to be 7 women that are the most vital to the social community. They seem to be the ones that are able to bring everyone together. Also, if they were to have conflicts, then the social network would severly be damaged.

Now we want to look at the projection of the events.

```python
projected_events = bipartite.projected_graph(G, clubs)
nx.draw(projected_events, with_labels=True)
plt.show()
```

So from looking at this graph, it appears that maybe event 8 and event 6 brought a lot of women together.

```python
nx.degree(projected_events)
```

There is something note worthy here. If in fact, these events are in chronological order, then we can assume that either events 6, 7, 8, and 9 were exceptionally cool, or that there was some influential women that influenced other women to attend these events. I imagine that up until event 6, the women were in 2 fairly disticnt groups. Then the main influencer from each group influenced their friends to go to event 6. At event 6, these friends became friends with members from the other group. This allowed for the next 3 events to have large gatherings as well. Then maybe something happened, or the friendships never really flourished causing the next events to have reduced attendence.

```python
eigen = nx.eigenvector_centrality(projected_events)
{k:v for k,v in sorted(eigen.items(), key=lambda item: item[1])}
```

Here we see that those 4 major events have the highest eigenvector centrality scores. We also see that the events that happened after these 4 events had higher scores than the ones that happend before these 4 events. What could this mean?  

```python
nx.betweenness_centrality(projected_events)
```

Here we see that events 6, 7, 8, and 9 are the only events that serve as bridges between other events.
