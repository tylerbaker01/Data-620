### The data

The data I have chosen to use is of a karate club from the 70's. Zachery was able to understand why the club fractured from this data.

```python
G = nx.karate_club_graph()
nx.draw(G, with_labels=True)
plt.show()
```

### Basic Analysis
#### Diameter
The diameter is the longest shortest path. The shortest path is simply the shortest path one can take from some node A to reach some other node B. So the diameter is essentially just stating what what two nodes are the furtherest away from each other and exactly how far they are. I will use networkx's built in function to calculate.

```python
nx.diameter(G)
```

#### Cycles
The cycles function returns all of the cycles within a graph. For example [8, 33, 31, 0] from a cycle. Meaning that if we start  with one of these numbers we can loop through the others to end at the node we started from.

```python
nx.cycle_basis(G)
```

### Centrality Measures
We weren't asked to find centrality measures but I would like to practice a few of them anyways.
#### Degree Centrality
This is the measure of how many edges are connected to a node.

```python
nx.degree_centrality(G)
```

This shows that nodes 33 and 0 are the most and second most popular nodes. I am going to guess that these 2 nodes represent the senseis that fracture.

#### Closeness Centrality
Closeness measures how close a node is to all other nodes in a graph.

```python
nx.closeness_centrality(G)
```

#### Betweenness Centrality
If a node has a high betweenness, then we can say that the node often acts as a bridge for other nodes.

```python
nx.betweenness_centrality(G)
```

This again shows that node 0 and node 33 are the highest.
