In this assignment, you’re asked to create the nodes and edges for a basic graph, such as the Krackhardt kite shown
below. (You’re welcome to substitute data of your own choosing).

```python
# Loading libraries
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

Create an empty graph.

```python
G = nx.Graph()
```

Produce nodes.

```python
nodes = ["A","B","C","D","E","F","G"]
G.add_nodes_from(nodes)
```

Produce edges.

```python
G.add_edges_from([("A","B"),("B","D"),("B","C"),("C","G"),("G","F"),("F","E"),("E","B"),("E","D"),("E","A")])
```

Display the graph.

```python
nx.draw(G, with_labels=True)
plt.show()
```
